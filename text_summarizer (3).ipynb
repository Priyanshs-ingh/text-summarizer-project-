```python
from tqdm import tqdm  # Ensure tqdm is imported if not already

def generate_batch_sized_chunks(list_of_elements, batch_size):
    """Split the dataset into smaller batches that can be processed simultaneously.
    Yield successive batch-sized chunks from list_of_elements."""
    for i in range(0, len(list_of_elements), batch_size):
        yield list_of_elements[i:i + batch_size]

def calculate_metric_on_test_ds(dataset, metric, model, tokenizer, batch_size=16, device='cpu', column_text="article", column_summary="highlights"):
    # Generate batches for articles and summaries
    article_batches = list(generate_batch_sized_chunks(dataset[column_text], batch_size))
    target_batches = list(generate_batch_sized_chunks(dataset[column_summary], batch_size))

    # Iterate through batches of articles and summaries
    for article_batch, target_batch in tqdm(zip(article_batches, target_batches), total=len(article_batches)):
        inputs = tokenizer(article_batch, max_length=1024, truncation=True, padding="max_length", return_tensors="pt")

        # Generate summaries
        summaries = model.generate(input_ids=inputs["input_ids"].to(device),
                                   attention_mask=inputs["attention_mask"].to(device),
                                   length_penalty=0.8, num_beams=8, max_length=128)

        # Decode generated summaries
        decoded_summaries = [tokenizer.decode(s, skip_special_tokens=True, clean_up_tokenization_spaces=True) for s in summaries]

        # Update the metric
        metric.add_batch(predictions=decoded_summaries, references=target_batch)

    # Compute and return the final score
    score = metric.compute()
    return score
```

```python
import torch
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
import fitz  # PyMuPDF

# Setup
nltk.download("punkt")
device = "cuda" if torch.cuda.is_available() else "cpu"
model_ckpt = "google/pegasus-cnn_dailymail"
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
model = AutoModelForSeq2SeqLM.from_pretrained(model_ckpt).to(device)

def summarize_pdf(pdf_path):
    """
    Extracts text from a PDF, adjusts it by word limits, and generates a summary using a pre-trained model.

    Args:
    - pdf_path (str): The file path to the PDF.

    Returns:
    - str: The summarized text of the PDF content in bullet points format.
    """

    # Step 1: Extract Text from PDF
    def extract_text_from_pdf(pdf_path):
        doc = fitz.open(pdf_path)
        text = ""
        for page in doc:
            text += page.get_text()
        doc.close()
        return text

    # Step 2: Adjust Text by Word Limits
    def adjust_text_by_word_limit(text, min_words, max_words):
        words = text.split()
        if len(words) < min_words:
            print("Text is too short for summarization.")
            return None
        if len(words) > max_words:
            return ' '.join(words[:max_words])
        return text

    # Step 3: Summarize Text to Bullet Points
    def summarize_to_bullets(text):
        max_chunk_length = 1024  # Set the max length for tokenizer
        words = word_tokenize(text)
        chunks = [' '.join(words[i:i+max_chunk_length]) for i in range(0, len(words), max_chunk_length)]

        summaries = []
        for chunk in chunks:
            inputs = tokenizer(chunk, return_tensors="pt", padding=True, truncation=True, max_length=max_chunk_length).to(device)
            summary_ids = model.generate(inputs["input_ids"], max_length=150, min_length=60, early_stopping=True)
            summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
            sentences = sent_tokenize(summary)
            bullet_summary = "\n".join(f"- {sentence}" for sentence in sentences)
            summaries.append(bullet_summary)

        return " ".join(summaries)

    # Extract and adjust text
    raw_text = extract_text_from_pdf(pdf_path)
    if not raw_text:
        return "No text extracted from PDF."

    adjusted_text = adjust_text_by_word_limit(raw_text, 300, 1000)  # Min 300 and max 1000 words
    if not adjusted_text:
        return "Text is too short or too long for summarization."

    # Generate summary
    return summarize_to_bullets(adjusted_text)

# Example usage
pdf_path = "/content/Service-NDA.pdf"  # Replace this with the path to your PDF
summary = summarize_pdf(pdf_path)
print(f"PDF Summary in Bullet Points:\n{summary}")
```

```python
import torch
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
import fitz  # PyMuPDF

# Setup
nltk.download("punkt")
device = "cuda" if torch.cuda.is_available() else "cpu"
model_ckpt = "google/pegasus-cnn_dailymail"
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
model = AutoModelForSeq2SeqLM.from_pretrained(model_ckpt).to(device)

def summarize_pdf(pdf_path):
    """
    Extracts text from a PDF, adjusts it by word limits, and generates a summary using a pre-trained model.

    Args:
    - pdf_path (str): The file path to the PDF.

    Returns:
    - str: The summarized text of the PDF content in bullet points format.
    """

    # Step 1: Extract Text from PDF
    def extract_text_from_pdf(pdf_path):
        doc = fitz.open(pdf_path)
        text = ""
        for page in doc:
            text += page.get_text()
        doc.close()
        return text.strip()

    # Step 2: Adjust Text by Word Limits
    def adjust_text_by_word_limit(text, min_words, max_words):
        words = text.split()
        if len(words) < min_words:
            print("Text is too short for summarization.")
            return None
        if len(words) > max_words:
            return ' '.join(words[:max_words])
        return text

    # Step 3: Summarize Text
    def summarize_text(text, max_length=1500, min_length=600, do_sample=False):
        inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=1024).to(device)
        summary_ids = model.generate(
            inputs["input_ids"],
            max_length=max_length,
            min_length=min_length,
            early_stopping=True,
            do_sample=do_sample
        )
        return tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    # Extract and adjust text
    raw_text = extract_text_from_pdf(pdf_path)
    if not raw_text:
        return "No text extracted from PDF."

    adjusted_text = adjust_text_by_word_limit(raw_text, 300, 1000)  # Min 300 and max 1000 words
    if not adjusted_text:
        return "Text is too short or too long for summarization."

    # Generate summary
    summary = summarize_text(adjusted_text)
    sentences = sent_tokenize(summary)
    bullet_points = "\n".join(f"- {sentence}" for sentence in sentences)
    return bullet_points

# Example usage
pdf_path = "/content/Service-NDA.pdf"  # Replace this with the path to your PDF
summary = summarize_pdf(pdf_path)
print(f"PDF Summary in Bullet Points:\n{summary}")
```

```python
import torch
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
import nltk
from nltk.tokenize import sent_tokenize
import fitz  # PyMuPDF
import streamlit as st

# Initialize tokenizer and model as None for lazy loading
tokenizer = None
model = None
device = "cuda" if torch.cuda.is_available() else "cpu"
model_ckpt = "google/pegasus-cnn_dailymail"

@st.cache(allow_output_mutation=True)
def load_model_and_tokenizer():
    global tokenizer, model
    if tokenizer is None or model is None:
        tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
        model = AutoModelForSeq2SeqLM.from_pretrained(model_ckpt).to(device)
    return tokenizer, model

def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    doc.close()
    return text.strip()

def adjust_text_by_word_limit(text, min_words, max_words):
    words = text.split()
    if len(words) < min_words:
        print("Text is too short for summarization.")
        return None
    if len(words) > max_words:
        return ' '.join(words[:max_words])
    return text

def summarize_text(text, max_length=1500, min_length=600, do_sample=False):
    tokenizer, model = load_model_and_tokenizer()
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=1024).to(device)
    summary_ids = model.generate(
        inputs["input_ids"],
        max_length=max_length,
        min_length=min_length,
        early_stopping=True,
        do_sample=do_sample
    )
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)

def summarize_pdf(pdf_path):
    raw_text = extract_text_from_pdf(pdf_path)
    if not raw_text:
        return "No text extracted from PDF."

    adjusted_text = adjust_text_by_word_limit(raw_text, 300, 1000)  # Min 300 and max 1000 words
    if not adjusted_text:
        return "Text is too short or too long for summarization."

    summary = summarize_text(adjusted_text)
    sentences = sent_tokenize(summary)
    bullet_points = "\n".join(f"- {sentence}" for sentence in sentences)
    return bullet_points

# Streamlit interface
st.title("PDF Summarizer")
pdf_file = st.file_uploader("Upload your PDF file", type=["pdf"])
if pdf_file:
    pdf_path = pdf_file.name
    with open(pdf_path, "wb") as f:
        f.write(pdf_file.getvalue())
    summary = summarize_pdf(pdf_path)
    st.text_area("Summary", summary, height=250)
```
